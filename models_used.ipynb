```python
# @title Year vs count()

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  counted = (series['Year']
                .value_counts()
              .reset_index(name='counts')
              .rename({'index': 'Year'}, axis=1)
              .sort_values('Year', ascending=True))
  xs = counted['Year']
  ys = counted['counts']
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Year', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Year')
_ = plt.ylabel('count()')
```

```python
#Print unique values in the 'Algorithm used' column
l1 = df['Algorithm used'].unique()
print(l1)
```

```python
algo_list = []
for i in l1:
  for j in i.split(','):
    algo_list.append(j)

print(algo_list)
```

```python
from collections import Counter
algorithm_counts = Counter(algo_list)
```

```python
algo_df = pd.DataFrame(algorithm_counts.items(), columns=["Algorithm","count"])
algo_df = algo_df.sort_values(by='count',ascending=False)
```

```python
#Top 10 algorithms

top_n = 10
top_algos = algo_df.head(top_n)


plt.figure(figsize=(20,10))
plt.bar(top_algos["Algorithm"],top_algos['count'],color='skyblue')
plt.xlabel("Algorithm")
plt.ylabel("Number of papers")
plt.title(f"Top {top_n} algorithms used in DDoS research")
plt.show()
```

```python
df = df.dropna(subset=['Accuracy'])
df['Accuracy'].isnull().sum()
```

```python
df['Accuracy']
```

```python
import re

'''
\b{re.escape(algo)}\b :
re.escape(algo) → Escapes special characters in the algorithm name (like + in K-Means++).
\b → Word boundary ensures the exact word is matched (avoiding partial matches).

.*?(\d+\.?\d*)% :
.*? → Matches any characters (lazy mode) until a number is found.
(\d+\.?\d*) → Captures accuracy value:
\d+ → One or more digits (e.g., 99).
\.? → Optional decimal point.
\d* → Optional decimal part (e.g., .5).
% → Ensures percentage format.
'''
def extract_accuracy(text):
    extracted = []
    for algo in algorithm_list:
        # Search for "<algo> ... <accuracy>%"
        match = re.search(rf'\b{re.escape(algo)}\b.*?(\d+\.?\d*)%', text)
        if match:
            extracted.append((algo, float(match.group(1))))
    return extracted if extracted else None

# Apply Function
df["Extracted Accuracy"] = df["Accuracy"].apply(extract_accuracy)

# Convert to a Structured DataFrame
df_exploded = df.explode("Extracted Accuracy").dropna().reset_index(drop=True)

# Separate into Algorithm and Accuracy columns
df_exploded["Algorithm"] = df_exploded["Extracted Accuracy"].apply(lambda x: x[0] if x else None)
df_exploded["Accuracy (%)"] = df_exploded["Extracted Accuracy"].apply(lambda x: x[1] if x else None)

# Drop the original extracted column
df_exploded = df_exploded.drop(columns=["Extracted Accuracy"])

df_exploded.tail()
```

```python
plt.figure(figsize=(10, 5))
plt.barh(df_exploded["Accurate_model"], df_exploded["model_accuracy"], color='skyblue')
plt.xlabel("Accuracy (%)")
plt.ylabel("Algorithm")
plt.title("Algorithm-wise Accuracy in DDoS Papers")
plt.show()
```

```python
all_features = []
for features_str in df['Imp feaures']:
    if isinstance(features_str, str):  # Check if the value is a string
      for feature in features_str.split(','):
          all_features.append(feature.strip())  # Remove whitespace

# Count feature occurrences
feature_counts = Counter(all_features)

# Convert to DataFrame for easier manipulation
features_df = pd.DataFrame.from_dict(feature_counts, orient='index', columns=['Count'])
features_df = features_df.sort_values('Count', ascending=False)
features_df = features_df.reset_index().rename(columns={"index": "Feature"})

# Display the top N features
top_n = 10  # Change as needed
print(features_df.head(top_n))


#Plot the top features
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.barh(features_df['Feature'].head(top_n), features_df['Count'].head(top_n), color='skyblue')
plt.xlabel("Number of papers")
plt.ylabel("Feature")
plt.title(f"Top {top_n} Features used in DDoS research")
plt.gca().invert_yaxis() # Invert y-axis to display the most frequent at the top
plt.show()
```
